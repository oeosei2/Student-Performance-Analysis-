```python
plt.hist(data['writing score'],bins=25)
plt.show()
```

```python
bins=[0,40,50,60,100]
math=pd.cut(data['math score'],bins=bins)
math.value_counts()
```

```python
sns.countplot(math,hue=data['gender'])
plt.xticks(rotation=90)
plt.show()
```

```python
bins=[0,40,60,80,100]
writing=pd.cut(data['writing score'],bins=bins)
writing.value_counts()
```

```python
sns.countplot(writing,hue=data['gender'])
plt.show()
```

```python
bins=[0,40,60,80,100]
reading=pd.cut(data['reading score'],bins=bins)
reading.value_counts()
```

```python
sns.countplot(reading,hue=data['gender'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(reading,hue=data['race/ethnicity'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(writing,hue=data['race/ethnicity'])
plt.show()
```
